### Q1f. Imputing Missing `hour_of_visit` Using Activity-Wise Mode

Claire wants to fill in the missing values of the `hour_of_visit` column.  
However, instead of filling all missing values with a single number, she wants to
impute them **based on activity category**:

- If a row has `activity == "grade"` and `hour_of_visit` is missing,  
  replace it with the **mode hour** of all `"grade"` activities.

- If a row has `activity == "visit"` and `hour_of_visit` is missing,  
  replace it with the **mode hour** of all `"visit"` activities.

- Similarly for `"submit"`.

**Steps to follow:**
- Create a copy of the original DataFrame named `portal_traffic_imputed`.
- Compute the mode hour for each activity category.
- Replace missing `hour_of_visit` values using the mode of that row‚Äôs activity.
- Ensure the resulting `hour_of_visit` column remains `np.float32`.


```python
import numpy as np
import pandas as pd

# Copy original DF
portal_traffic_imputed = portal_traffic.copy()

# Step 1: Compute activity-wise mode for hour_of_visit
activity_modes = (
    portal_traffic.groupby("activity")["hour_of_visit"]
    .agg(lambda x: x.mode().iloc[0])      # get mode value
    .astype(np.float32)                   # ensure float32
)

# Step 2: Impute missing values using mode per activity
portal_traffic_imputed["hour_of_visit"] = (
    portal_traffic_imputed
        .apply(lambda row: activity_modes[row["activity"]]
               if pd.isna(row["hour_of_visit"]) else row["hour_of_visit"],
               axis=1)
).astype(np.float32)

portal_traffic_imputed
```

- Store the result in a DataFrame named `sorted_merged_info`.
- Add a new column `last_letter` containing the **last letter of the participant‚Äôs
