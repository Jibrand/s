### Filtering with boolean arrays

Filtering is the process of removing unwanted entries. In your quest for cleaner data, you will undoubtedly filter your data at some point: whether it be for clearing up cases with missing values, for culling out fishy outliers, or for analyzing subgroups of your dataset.

For your reference, some commonly used comparison operators are given below.

Symbol | Usage      | Meaning 
------ | ---------- | -------------------------------------
==   | a == b   | Does a equal b?
<=   | a <= b   | Is a less than or equal to b?
&gt;=   | a >= b   | Is a greater than or equal to b?
<    | a < b    | Is a less than b?
&#62;    | a &#62; b    | Is a greater than b?
~    | ~p       | Returns negation of p
&#124; | p &#124; q | p OR q
&    | p & q    | p AND q
^  | p ^ q | p XOR q (exclusive or)

### Olympic Medals Dataset

```python
olympics = pd.read_csv("olympic.csv")
olympics.head()
```

### Question 6a: Find the maximum medals per country using groupby.agg.

```python
# max_medals = olympics.groupby("Country")["Medal"].agg("count").sort_values(ascending=False)

max_medals = olympics.groupby("Sport")["Medal"].agg("count").sort_values(ascending=False)

max_medals.head()
```

### Question 6b: Show all details for each country's best performance. First find the idx followed by best_results

```python
idx = olympics.groupby("Country").idxmax()
best_results  = olympics.groupby("Country").agg("count")
print(idx.head())
print(best_results.head(5))
```

```python
grader.check("q6b")
```

### Question 6C: Filter results after 2010 for countries with at least 10 medals in any year.

```python
post_2010 = olympics.loc[olympics["Year"].astype(int) > 2010]

medal_counts = post_2010[["Country", "Year", "Medal"]].groupby(["Country", "Year"]).agg(
    MedalCount=("Medal", "count")
    )

major_countries  = medal_counts[medal_counts["MedalCount"] >= 10].reset_index()

major_countries
```
